```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
```

```python
def cost(x ,y,m):
    ypred=(m*x).sum(axis=1)
    diff=(y-ypred)**2
    return diff.mean()/2;

def coefficient(x,y,m):
    ypred=(m*x).sum(axis=1)
    u=(y-ypred)**2
    v=(y-(y.mean()))**2
    return 1-(u.sum()/v.sum())

    
```

```python
def batch_grad(m,x,y,r):
    col=x.shape[1]
    temp_m=m.copy()
    for i in range(col):
        temp=x.iloc[:,i]*y-(((temp_m*x)).sum(axis=1))*x.iloc[:,i]
        m[i]=m[i]+(r*(temp.mean()))
        #print(r*temp.mean(),m[i])
        #break
    return m
```

```python
def gradientdescent(xtrain,ytrain):
    col=xtrain.shape[1]
    m=np.random.rand(xtrain.shape[1])
    itr=0
    rate=0.01
    temp_c=2000
    while(temp_c>1 and itr<3000):
        itr+=1
        if(itr==1600 or itr==4500 or itr==6500 or itr==13000):
            rate*=0.1
        xtrain,ytrain=shuffle(xtrain,ytrain)
        m=batch_grad(m,xtrain.iloc[:300,:],ytrain.loc[:300],rate)
        temp_c=cost(xtrain,ytrain,m)
        print(itr,temp_c)
    return m
```

```python
data=pd.read_csv('training_boston_train.csv')
xtrain=data.iloc[:,:-1]
col=xtrain.shape[1]
for i in range(col):
    for j in range(i,col):
        xtrain[xtrain.columns[i]+'_'+xtrain.columns[j]]=xtrain.iloc[:,i]*xtrain.iloc[:,j]
scaler=StandardScaler()
columns=xtrain.columns
xtrain=scaler.fit_transform(xtrain)
xtrain=pd.DataFrame(xtrain,columns=columns)
xtrain['constant']=1
ytrain=data.iloc[:,-1]
```

```python

m=gradientdescent(xtrain,ytrain)
```

```python
print(coefficient(xtrain,ytrain,m))
data=pd.read_csv('test_boston_test.csv',names=xtrain.columns[:13])
print(data.shape)
col=data.shape[1]
for i in range(col):
    for j in range(i,col):
        data[data.columns[i]+'_'+data.columns[j]]=data.iloc[:,i]*data.iloc[:,j]
data=scaler.transform(data)
data=pd.DataFrame(data,columns=columns)
data['constant']=1
result=(data*m).sum(axis=1)
result
```
